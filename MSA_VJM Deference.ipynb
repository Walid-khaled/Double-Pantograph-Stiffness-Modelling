```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
MSA_y = np.loadtxt('MSA_deflection.txt')
VJM_y = np.loadtxt('VJM_deflection.txt')
Average = [(MSA_y[i]+VJM_y[i])/(2) for i in range(len(MSA_y))]
Deference = [(np.absolute(MSA_y[i]-VJM_y[i])) for i in range(len(Average))]
dz_limit = [0.9]
z_pos = np.linspace(0.1, dz_limit, 30)    
```

```python
plt.scatter(z_pos, MSA_y, label='MSA_y')
plt.scatter(z_pos, VJM_y, color = 'red', label='VJM_y')
plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
plt.plot(z_pos, Average, '--k', label='Average')
plt.ylabel('Deflection \u0394t (m)',fontsize=20)
plt.xlabel('z',fontsize=20)
plt.legend()
plt.savefig('Deflection Average.png')
plt.show()
files.download('Deflection Average.png')
```
